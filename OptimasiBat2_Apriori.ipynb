```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Membaca file CSV dan mengonversinya ke dalam DataFrame
data = pd.read_csv('datapenjualan1.csv')

# Mengonversi struktur DataFrame yang dibaca ke format yang diinginkan
df_transactions = data.groupby('TransactionID')['Items'].apply(list).reset_index()

df_transactions = pd.DataFrame(data)

# Fungsi untuk menghasilkan DataFrame biner dari transaksi
def create_binary_dataframe(transactions):
    unique_items = list(set(item for sublist in transactions for item in sublist))
    binary_dataframe = pd.DataFrame(0, columns=unique_items, index=range(len(transactions)))

    for i, transaction in enumerate(transactions):
        binary_dataframe.loc[i, transaction] = 1

    return binary_dataframe.astype(bool)  # Convert DataFrame to boolean

# Fungsi objektif yang akan dioptimalkan
def objective_function(parameters):
    # Parameters: min_support dan min_confidence
    min_support = parameters[0]
    min_confidence = parameters[1]

    # Convert transaksi menjadi DataFrame biner
    binary_dataframe = create_binary_dataframe(df_transactions['Items'])

    # Implementasi Algoritma Apriori dengan parameter yang diberikan
    frequent_itemsets = apriori(binary_dataframe, min_support=min_support, use_colnames=True)
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)

    # Menghitung evaluasi berupa total nilai Lift
    total_lift = rules['lift'].sum()

    return total_lift

# Implementasi Bat Algorithm untuk optimasi
def bat_algorithm(num_bats, num_iterations):
    # Inisialisasi parameter Bat Algorithm
    num_dimensions = 2  # Jumlah parameter yang ingin dioptimalkan (min_support, min_confidence)
    lower_bound = np.array([0.01, 0])  # Batas bawah untuk setiap parameter
    upper_bound = np.array([1, 1])  # Batas atas untuk setiap parameter
    loudness = 0.5  # Inisialisasi nilai loudness
    pulse_rate = 0.5  # Inisialisasi nilai pulse rate

    # Inisialisasi posisi kelelawar secara acak di dalam batas pencarian
    bats = np.random.uniform(lower_bound, upper_bound, (num_bats, num_dimensions))
    # Inisialisasi kecepatan dan posisi awal
    velocities = np.zeros((num_bats, num_dimensions))
    frequencies = np.zeros(num_bats)

    # Inisialisasi posisi terbaik
    best_solution_index = 0
    best_solution = bats[best_solution_index]

    # Iterasi Bat Algorithm
    for _ in range(num_iterations):
        # Update posisi kelelawar
        for i in range(num_bats):
            # Pembaruan frekuensi dan posisi berdasarkan pulsasi
            if np.random.random() > frequencies[i]:
                velocities[i] += (bats[i] - bats.mean(axis=0)) * loudness
                frequencies[i] = pulse_rate * (1 - np.exp(-1 * np.random.random()))  # Update frekuensi

            # Batasan posisi kelelawar dalam batas pencarian
            bats[i] = np.clip(bats[i] + velocities[i], lower_bound, upper_bound)

        # Evaluasi setiap kelelawar
        for i in range(num_bats):
            current_evaluation = objective_function(bats[i])
            # Jika nilai evaluasi lebih baik, update posisi terbaik
            if current_evaluation > objective_function(best_solution):
                best_solution = bats[i]

    # Mengembalikan posisi terbaik (parameter yang dioptimalkan)
    return best_solution

# Menggunakan Bat Algorithm untuk mencari parameter terbaik
best_parameters = bat_algorithm(10, 100)
print("Best Parameters (min_support, min_confidence):", best_parameters)
```
